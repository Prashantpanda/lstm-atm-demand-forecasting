```python
import pandas as pd
import numpy as np
from numpy import array

import tensorflow as tf
import keras
import matplotlib.pyplot as plt
# from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras import optimizers

from IPython.display import display
```

```python
data =  pd.read_csv("AggregatedData.csv", low_memory=False)
pd.options.display.max_columns = None
```

```python
# consider only Big Street ATM
data = data[data["ATM Name"] == 'Big Street ATM']

# replace inconsistent Weekday values

data['Weekday'].replace('MONDAY', 'Monday', inplace=True)
data['Weekday'].replace('TUESDAY', 'Tuesday', inplace=True)
data['Weekday'].replace('WEDNESDAY', 'Wednesday', inplace=True)
data['Weekday'].replace('THURSDAY', 'Thursday', inplace=True)
data['Weekday'].replace('FRIDAY', 'Friday', inplace=True)
data['Weekday'].replace('SATURDAY', 'Saturday', inplace=True)
data['Weekday'].replace('SUNDAY', 'Sunday', inplace=True)

# handle some inconsistent date strings and convert to datetime format
contain_slash = data[data['Transaction Date'].str.contains('/')].copy()
not_contain_slash = data[~data['Transaction Date'].str.contains('/')].copy()

contain_slash['Transaction Date'] = pd.to_datetime(contain_slash['Transaction Date'], format = "%d/%m/%Y")
not_contain_slash['Transaction Date'] = pd.to_datetime(not_contain_slash['Transaction Date'])

data = contain_slash.append(not_contain_slash)


# set index of data to be the transaction date and drop the transaction date column 
data.index = data['Transaction Date']
data = data.drop(['Transaction Date'], axis=1)

# extract year and month from datetime
# data['Year'] = data.index.year
# data['Month'] = data.index.month

# drop columns that won't be used as predictors later on
data = data.drop(['ATM Name','No Of Withdrawals','No Of XYZ Card Withdrawals', 'No Of Other Card Withdrawals','Amount withdrawn XYZ Card','Amount withdrawn Other Card'], axis=1)

data = data.sort_index()

```

```python
# find date range
min_date = min(data.index)
max_date = max(data.index)

# find missing date values
pd.date_range(start = min_date, end = max_date).difference(data.index)

# not handling missing dates for now
```

```python
# hot encode categorical values
final_dataset = pd.get_dummies(data, columns = ['Weekday','Festival Religion', 'Working Day', 'Holiday Sequence'])
```

```python
final_dataset.head()
```

```python
# ensure data type is float
input_values = final_dataset.drop('Total amount Withdrawn', axis=1).values
input_values = input_values.astype('float32')

output_values = final_dataset['Total amount Withdrawn'].values
output_values = output_values.astype('float32')
```

```python
# normalize features
# scaler = MinMaxScaler(feature_range=(0, 1))
# scaled = scaler.fit_transform(input_values)
```

```python
print(input_values.shape)
print(output_values.shape)
```

```python
# steps = 2
# dim_0 = test_input.shape[0] - steps
# dim_1 = test_input.shape[1]

# x = np.zeros((dim_0, steps, dim_1))
# y = np.zeros((dim_0,))

# for i in range(dim_0):
#     x[i] = test_input[i : steps + i]
#     y[i] = test_output[0]
# print("length of x and y",x.shape,y.shape)

def shape(steps, inputs, outputs):
    dim_0 = inputs.shape[0] - steps
    dim_1 = inputs.shape[1]
    
    x = np.zeros((dim_0, steps, dim_1))
    y = np.zeros((dim_0,))
    
    for i in range(dim_0):
        x[i] = inputs[i : steps + i]
        y[i] = outputs[i + steps]
        
    return x,y
```

```python
train_X, train_y = shape(0,input_values,output_values)
```

```python
train_X
```

# LSTM

```python
model = Sequential()
model.add(LSTM(50, input_shape=(train_X.shape[1], train_X.shape[2])))
model.add(Dense(1))
# model.compile(loss='mae', optimizer='adam')
model.compile(loss="mean_squared_error", optimizer = "adam")
# fit network
history = model.fit(train_X, train_y, epochs=500, batch_size=10, verbose=2, shuffle=False)
```

```python
# # plot history
plt.plot(history.history['loss'], label='train')
# pyplot.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```

```python
pred_y = model.predict(train_X)
pred_y
```
